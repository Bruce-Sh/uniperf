```python
import re
def parse_sfssum(log, suffix):
    print("Got SFS log file {} and suffix {}".format(log, suffix))
    with open(log, 'r') as f:
        for line in f.readlines():
           
           s1 = re.search(r'^\d+', line)
           if(s1):
                
                entry = re.split(r',', line.strip())
                print(len(entry))
                
```
